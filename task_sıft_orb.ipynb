```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,8)
```

```python
image_path="download.jpeg"
image=cv2.imread(image_path)
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
fast=cv2.FastFeatureDetector_create()
keypoints=fast.detect(gray,None)

image_default=cv2.drawKeypoints(image_rgb,keypoints,None,color=(0,255,0))
```

```python
print(f"eşik değer ",fast.getThreshold())
print(f"nonmaxSuppression ",fast.getNonmaxSuppression())
print(f"toplam keypoint ",fast.getType())
print(f"keypoints sayısı ",len(keypoints))
```

```python
plt.imshow(image_default)
plt.axis("off")
plt.show()
```
